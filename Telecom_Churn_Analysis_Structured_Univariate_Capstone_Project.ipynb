**UNIVARIATE ANALYSIS**

```python
# NO OF UNIQUE STATES:
df['State'].nunique()
```

```python
state=df['State'].value_counts()
state.plot(kind='bar')
plt.rcParams['figure.figsize'] = (40,5)
plt.title('Customers Distribution Across States')
plt.ylabel('Number of customers')
plt.xlabel('States')
```
